```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
n_nf = np.zeros((8))
diam_nf = np.zeros((8))
cc_nf = np.zeros((8))

for t in range(1,4):
    path = 'csv_files/Erdos-Renyi_diam_cc.csv'
    with open(path) as file:
        csv_reader = csv.DictReader(file)
        i=0
        for row in csv_reader:
            n_nf[i] = row["n"]
            diam_nf[i] = row["diameter"]
            cc_nf[i] = row["clustering"]
            i+=1
```

```python
degree_nf = np.zeros((24))
num_nf = np.zeros((24))
# cc_nf = np.zeros((8))

for t in range(1,4):
    path = 'csv_files/Erdos-Renyi_dd_1000.csv'
    with open(path) as file:
        csv_reader = csv.DictReader(file)
        i=0
        for row in csv_reader:
            degree_nf[i] = row["degree"]
            num_nf[i] = row["num"]
            i+=1
```

```python
fig,ax = plt.subplots()
plt.semilogy(n_nf,diam_nf, '.', markersize = 10, color='blue', label='NF')
m, b = np.polyfit(np.log(n_nf),np.log(diam_nf), 1)
y = np.exp(m*np.log(n_nf)+b)
plt.semilogy(n_nf, y, color='blue')
ax.set_xlabel('Number of Items')
ax.set_ylabel('Diameter')
ax.set_title('N vs. Diameter for Erdos-Renyi (lin-log)')
plt.show()
```

```python
fig,ax = plt.subplots()
plt.semilogy(n_nf,cc_nf, '.', markersize = 10, color='blue', label='NF')
m, b = np.polyfit(np.log(n_nf),np.log(cc_nf), 1)
y = np.exp(m*np.log(n_nf)+b)
plt.semilogy(n_nf, y, color='blue')
ax.set_xlabel('Number of Items')
ax.set_ylabel('Diameter')
ax.set_title('N vs. Clustering Coefficient for Erdos-Renyi (lin-log)')
plt.show()
```

```python
fig,ax = plt.subplots()
plt.loglog(n_nf,cc_nf, '.', basex=2, basey=2, markersize = 10, color='blue', label='NF')
m, b = np.polyfit(np.log(n_nf),np.log(cc_nf), 1)
y = np.exp(m*np.log(n_nf)+b)
plt.loglog(n_nf, y, basex=2, basey=2, color='blue')
ax.set_xlabel('Number of Items')
ax.set_ylabel('Clustering Coefficient')
ax.set_title('N vs. Clustering Coefficient for Erdos-Renyi (log-log)')
plt.show()
```

```python
fig,ax = plt.subplots()
plt.loglog(degree_nf,num_nf, '.', basex=2, basey=2, markersize = 10, color='blue', label='NF')
m, b = np.polyfit(np.log(degree_nf),np.log(num_nf), 1)
y = np.exp(m*np.log(degree_nf)+b)
plt.loglog(degree_nf, y, basex=2, basey=2, color='blue')
ax.set_xlabel('Degree of Vertex')
ax.set_ylabel('number of vertices with the degree')
ax.set_title('Degree vs Number of vertices with the degree (log-log)')
plt.show()
```

```python
fig,ax = plt.subplots()
plt.plot(degree_nf,num_nf, '.', markersize = 10, color='blue', label='NF')
m, b = np.polyfit(np.log(degree_nf),np.log(num_nf), 1)
y = m* degree_nf +b
plt.plot(degree_nf, y, color='blue')
ax.set_xlabel('Degree of Vertex')
ax.set_ylabel('number of vertices with the degree')
ax.set_title('Degree vs Number of vertices with the degree (lin-lin)')
plt.show()
```
